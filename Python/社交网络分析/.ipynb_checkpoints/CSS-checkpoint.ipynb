```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
from dateutil.parser import parse
```

```python
plt.style.use('ggplot')


dateparse = lambda dates: pd.datetime.strptime(
    dates, '%Y-%m-%d'
)
d1=pd.read_csv('movie_Boxoffice.csv',parse_dates=['ReleaseTime'],date_parser=dateparse)
d1 = d1.set_index('ReleaseTime').iloc[:,1:]
# d1.BoxOffice.apply(lambda x: x/100)
d2=pd.read_csv('movie_person.csv')

f1=plt.figure(dpi=200)
d1.resample('Q').count().to_period('Q').MovieID.plot(kind='bar',fontsize=6)
plt.xlabel('各季度',fontsize=8)
plt.ylabel('上映数量',fontsize=8)

f2=plt.figure(dpi=200)
ax1=plt.gca()
dd2=d1.loc[:,['MovieName','AvgPeople','BoxOffice']].sort_values(by='BoxOffice',ascending=False)
ax1.bar(list(dd2.MovieName),list(dd2.AvgPeople))
plt.xticks(rotation=90,fontsize=3)

# ax2=ax1.twinx
# dd2=d1.loc[:,['MovieName','BoxOffice']].set_index('MovieName').sort_values(by='BoxOffice',ascending=False).plot(fontsize=6,ax=ax2)
# ax1.bar(dd2.MovieName,dd2.BoxOffice)


plt.show()
```
