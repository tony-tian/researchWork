```python
import requests,json,re
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from bs4 import BeautifulSoup

import pymysql.cursors
```

```python
def getPiaoFang():
    url=r'http://www.cbooo.cn/Alltimedomestic'
    rurl=requests.get(url).text
    soup = BeautifulSoup(rurl,'lxml')  
    trs = soup.find_all('tr',{"class":"date date01"})  
    print(trs)
    ulist=[]
    for tr in trs:  
        ui = []  
        for td in tr:  
            ui.append(td.string)  
        ulist.append(ui)  
    print(ulist)
    
getPiaoFang()    
```

```python
from tqdm import tnrange,tqdm_notebook
d2=pd.DataFrame({'movie':[],'actor':[]})
for i in tqdm_notebook(range(5),desc='1st loop'):    
    d1=pd.read_json(requests.get('http://www.cbooo.cn/BoxOffice/getInland?pIndex='+str(i+1)+'&t=0').text)
    p=re.compile(r'<a target="_blank".*?title="(.*?\ .+?)">')
    for item in d1.iterrows():
        m_list=p.findall(requests.get('http://www.cbooo.cn/m/'+str(item[1][6])).text)
        a=[item[1][7]]*len(m_list)
        d2=pd.concat([d2,pd.DataFrame({'movie':a,'actor':m_list})])
    #     d2.append(({'movie':a,'actor':m_list}),ignore_index=True)
d2.reset_index(drop=True)
```
